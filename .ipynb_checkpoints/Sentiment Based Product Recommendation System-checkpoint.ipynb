```python
#--- HTML Tag Removal
#---- Function to remove html tags from data
def striphtml(data):
    p = re.compile('<.*?>') #Find this kind of pattern
    return p.sub('', data) # Substitute nothing at the place of strings which matched the patterns
```

```python
#--- Punctuation removal
#---- Function to remove all the punctuation from the text
def strippunc(data):
    p = re.compile(r'[?|!|\'|"|#|.|,|)|(|\|/|~|%|*]')
    return p.sub('', data)
```

```python
#--- Stopwords
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
stop = stopwords.words('english')
from nltk.stem import SnowballStemmer
```

```python
snow = SnowballStemmer('english')

def nltk_tag_to_wordnet_tag(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV    
    else:
        return None
```

```python
def lemmatize_sentence(sentence):
    # tokenize the sentence and find the POS tag for each token
    nltk_tagged = nltk.pos_tag(nltk.word_tokenize(sentence))
    #tuple of (token, wordnet_tag)
    wordnet_tagged = map(lambda x: (x[0], nltk_tag_to_wordnet_tag(x[1])), nltk_tagged)
    lemmatized_sentence = []
    for word, tag in wordnet_tagged:
        if tag is None:
            #if there is no available tag, append the token as is
            lemmatized_sentence.append(snow.stem(word)) #stem the word if no lemma is obtaines
        else:
            #else use the tag to lemmatize the token
            lemmatized_sentence.append(lemmatizer.lemmatize(word, tag))
    return " ".join(lemmatized_sentence)
```

```python
#pre-processing output for one Review
def preprocessText(text, stem=False):
    text=text.lower()
    text = striphtml(text)
    text = strippunc(text)
    return lemmatize_sentence(text)
```

### Find the top ngram frequency among the Reviews

```python
#Write code to find the top ngram frequency among the Reviews
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from pprint import pprint

def get_top_ngrams(text, n=None, ngram=(1,1)):
    vect = CountVectorizer(stop_words='english', ngram_range=ngram).fit(text)
    bag_of_words = vect.transform(text)
    sum_words = bag_of_words.sum(axis=0)
    words_frequency = [(word, sum_words[0, index]) for word, index in vect.vocabulary_.items()]
    words_frequency = sorted(words_frequency, key = lambda x: x[1], reverse=True)
    return words_frequency[:n]
```

### Print the top 30 words in the unigram frequency

```python
top_30words = get_top_ngrams(df_main['Review'].values.astype('U'), n=30, ngram=(1,1))
df_unigram = pd.DataFrame(top_30words, columns=['unigram', 'count'])
df_unigram
```
